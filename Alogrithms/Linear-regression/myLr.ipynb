```python
class MyLR:

    def __init__(self):
        self.m= None
        self.b= None


    def fit(self,X_train,Y_train):
        numerator=0
        denominator=0

        for i in range(X_train.shape[0]):
            numerator += ((X_train.iloc[i] - X_train.mean()) * (Y_train.iloc[i] - Y_train.mean()))

            denominator += ((X_train.iloc[i] - X_train.mean()) ** 2)

        self.m=numerator/denominator     
        self.b= Y_train.mean() - self.m*X_train.mean()

        print(self.m)
        print(self.b)

    def predict(self,X_test):
        return self.m*X_test+self.b    
```

```python
from kagglehub import KaggleDatasetAdapter
import kagglehub
from dotenv import load_dotenv

# Note the correct file name here
file_path = "disease_prediction.csv"
load_dotenv()

# load_dataset is deprecated, but you can still use it or switch to kagglehub.dataset_load
df = kagglehub.dataset_load(
    KaggleDatasetAdapter.PANDAS,
    "shyamnadhs/heart-disease-prediction-dataset",
    file_path,
)

# Test it
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()


df['disease']=le.fit_transform(df['disease'])





```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df[['heart_rate']],df['disease'],test_size=0.2,random_state=42)



model=MyLR()

model.fit(X_train,Y_train)

model.predict(0.454)
```
